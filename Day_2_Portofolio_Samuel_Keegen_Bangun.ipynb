## Filling missing value

```python
gold_rev = gold.copy()
```

gold_rev sebagai dataframe baru untuk mengecek 'apakah missing value yang diisi dengan O dapat membuat data menjadi lebih baik distribusinya' 

```python
# Fill NonNull with '0'
gold_rev['China(CNY)'] = gold['China(CNY)'].fillna(0)
gold_rev['Turkey(TRY)'] = gold['Turkey(TRY)'].fillna(0)
gold_rev['Vietnam(VND)'] = gold['Vietnam(VND)'].fillna(0)
gold_rev['Egypt(EGP)'] = gold['Egypt(EGP)'].fillna(0)
```

- Nilai missing value diisikan dengan values 0 untuk mengecek apakah datanya semakin terdistribusi normal

```python
gold_rev[gold_rev['China(CNY)'] == 0]
```

- Diatas merupakan contoh beberapa rows missing values yang sudah diisi dengan 0 pada kolom China(CNY)

```python
# Description of gold-rev
gold_rev.describe()
```

- Bila dibandingkan mean dan median/50% dari dataframe gold dan gold_rev dapat dilihat bahwa dengan mengisi missing value dengan 0 membuat data semakin tidak terdistribusi normal. 
- Selain itu, sangat tidak wajar jika harga emas pada suatu waktu di suatu negara diisi dengan 0. Begitu pula, apabila diisi dengan mean, median maupun modus. Hal itu karena, nilai emas tidak mungkin lagi 0 pada tahun missing value tersebut dan tidak boleh diisi sembarangan angka, mengingat harga emas melibatkan berbagai faktor dalam penentuannya seperti kesehatan ekonomi suatu negara dan faktor-faktor lainnya.
- Oleh karena itu, akan lebih baik bila data yang missing di drop saja.

# Transforming Date

```python
# Transform string to date
gold['Date'] = pd.to_datetime(gold['Date'], format='%d-%m-%Y')
```

```python
# Gold dataframe info
gold.info()
```

- Setelah dilakukannya Tranform Data pada kolom Date diperoleh kolom Date dengan jenis data datetime

```python
# Extracting Year, Month, Days
gold['Year'] = gold['Date'].dt.year
gold['Month'] = gold['Date'].dt.month
gold['Day'] = gold['Date'].dt.day
```

```python
# Gold dataframe info
gold.info()
```

- Dengan dilakukannya extracting diperoleh kolom baru yaitu Year, Month dan Day merupakan pecahan dari value kolom Date
- Ketiga kolom baru tersebut merupakan jenis data Integer

```python
gold.head()
```

```python
# Categorize month
def month_cat(x):
    if x == 1:
        return 'Jan'
    elif x == 2:
        return 'Feb'
    elif x == 3:
        return 'Mar'
    elif x == 4:
        return 'Apr'
    elif x == 5:
        return 'May'
    elif x == 6:
        return 'Jun'
    elif x == 7:
        return 'Jul'
    elif x == 8:
        return 'Aug'
    elif x == 9:
        return 'Sep'
    elif x == 10:
        return 'Oct'
    elif x == 11:
        return 'Nov'
    elif x == 12:
        return 'Des'

gold['MonthCategory'] = gold['Month'].apply(month_cat)
```

Hal ini ditujukan untuk mempermudah visulisasi data

```python
# Top 10 Highest Gold Price in United States(USD)
Top10_US = (gold
                 .groupby(['Year', 'MonthCategory'])
                 .agg(Price = ('United States(USD)', 'max'))
                 .reset_index()
                 .sort_values('Price', ascending=False)
                 .head(10))
Top10_US
```

- Dalam United States(USD) harga emas tertinggi ialah pada Tahun 2020 bulan Juli. Hal ini perlu diingat terlepas dari data yang sudah di drop

- Dalam United States(USD) harga emas terendah ialah pada Tahun 1999 bulan Agustus. Hal ini perlu diingat terlepas dari data yang sudah di drop
